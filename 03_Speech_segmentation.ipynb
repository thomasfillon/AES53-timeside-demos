```python
from __future__ import division

import timeside

wavFile = '/home/thomas/code/timeside/exempleCMMR/CNRSMH_I_2013_202_001_06.mp3'

decoder = timeside.decoder.FileDecoder(wavFile, start=0, duration = 60)
decoder.output_blocksize=2048

irit4hz = timeside.analyzer.IRITSpeech4Hz()

specgram = timeside.analyzer.Spectrogram()

iritEntropy = timeside.analyzer.IRITSpeechEntropy()
```

```python
# Define and run the pipe in one line !
(decoder | irit4hz | iritEntropy | specgram).run()
```

```python
# Display avalaible results
irit4hz.results.keys()
```

```python
# Pick segmentation results
irit_4hz_seg = irit4hz.results['irit_speech_4hz.segments']
```

```python
# Pick segmentation results
irit_entropy_seg = iritEntropy.results['irit_speech_entropy.segments']
```
